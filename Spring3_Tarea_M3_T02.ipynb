**Observaciones:**

ActualElapsedTime, AirTime y ArrDelay con 885 registros de NaN,  TaxiIn con 740 registros, y TaxiOut con 50 y CRSElapsedTime con 21 regitros.

* Con respectos a las variables ActualElapsedTime, AirTime, TaxiIn, TaxiOut y CRSElapsedTime, como son variables que se encuentran en función del tiempo serán sustituidas por su media (mean).

```python
#En las causas y el tiempo de retrasado de los arribos del avion, sustituir los valores NaN con el valor 0.0.
new_data2008['WeatherDelay'] = new_data2008['WeatherDelay'].fillna(0)
new_data2008['SecurityDelay'] = new_data2008['SecurityDelay'].fillna(0)
new_data2008['ArrDelay'] = new_data2008['ArrDelay'].fillna(0)
new_data2008['CarrierDelay'] = new_data2008['CarrierDelay'].fillna(0)
new_data2008['NASDelay'] = new_data2008['NASDelay'].fillna(0)
new_data2008['LateAircraftDelay'] = new_data2008['LateAircraftDelay'].fillna(0)

# En los valores faltantes que las variables se encuentran en función del tiempo, sustituir por la media.
ActualElapsedTime_mean = new_data2008['ActualElapsedTime'].mean()
new_data2008['ActualElapsedTime'] = new_data2008['ActualElapsedTime'].fillna(ActualElapsedTime_mean)
AirTime_mean = new_data2008['AirTime'].mean()
new_data2008['AirTime'] = new_data2008['AirTime'].fillna(AirTime_mean)
TaxiIn_mean = new_data2008['TaxiIn'].mean()
new_data2008['TaxiIn'] = new_data2008['TaxiIn'].fillna(TaxiIn_mean)
TaxiOut_mean = new_data2008['TaxiOut'].mean()
new_data2008['TaxiOut'] = new_data2008['TaxiOut'].fillna(TaxiOut_mean)
CRSElapsedTime_mean = new_data2008['CRSElapsedTime'].mean()
new_data2008['CRSElapsedTime'] = new_data2008['CRSElapsedTime'].fillna(CRSElapsedTime_mean)

new_data2008
```

```python
new_data2008.isnull().sum().sort_values(ascending = False)
```

* **Crear nuevas columnas**
    * Velocidad media del vuelo. 
    * Si ha llegado tarde o no. 
    * Salido tarde o no.
    * Total del retraso.

```python
#Calculo de la velocidad media con las variables (columnas) 'Distance' y 'AirTime'.

new_data2008['flight_speed'] = new_data2008['Distance'] / (new_data2008['AirTime']/60)
new_data2008.head().reset_index()
```

```python
#Identificar si esta retrasado en salida o llegada

def delay_interval(x): 
    if x <= 15: 
        mensaje = "Ontime"
    else: 
        mensaje = "Delay"
    return mensaje

new_data2008['Is_ArrDelay'] = new_data2008['ArrDelay'].apply(delay_interval)
                            
new_data2008['Is_DepDelay'] = new_data2008['DepDelay'].apply(delay_interval)

new_data2008.head().reset_index()
```

```python
# Suma total de los retrasos (salida + llegada)
new_data2008["Total_Delay"] = new_data2008["ArrDelay"] + new_data2008["DepDelay"]
new_data = new_data2008
new_data
```

* **Haz una tabla de las aerolíneas con mayores retrasos acumulados.**
